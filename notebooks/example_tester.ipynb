```python
import torch
from src.data_factory import DataFactory
from src.model import BasicTransformer
from src.tester import CapacityTester
from src.trainer import MemorizationTrainer
```

```python
vocab_size = 8
prefix_len = 5
payload_len = 5
payload_seed = 0
d_model = 32
n_heads = 4
ffn_multiplier = 4
n_layers = 1
```

```python
dataset_factory = DataFactory()
dataset = dataset_factory.get_dataset(vocab_size=vocab_size, prefix_len=prefix_len, payload_len=payload_len, payload_seed=1, shuffle_seed=1)
print(dataset)
```

```python
if torch.backends.mps.is_available():
    device = "mps"
    # torch.use_deterministic_algorithms(True)  # uncomment to enable deterministic behavior
    print("Using MPS acceleration.")
else:
    device = "cpu"
    print("Using CPU.")
```

```python
payload_seed = 1000
shuffle_seed = 2000
init_seed = 3000

def build_fresh_trainer():
    torch.manual_seed(init_seed)

    model = BasicTransformer(vocab_size, d_model, n_heads, ffn_multiplier, n_layers).to(device)
    return MemorizationTrainer(
        model=model,
        prefix_len=prefix_len,
        device=device
    )

tester = CapacityTester(
    make_trainer_fn=build_fresh_trainer,
    dataset=dataset,
    start_n=128,
    search_tolerance=5,
    batch_size=128,
    max_epochs=500,
    init_seed=init_seed
)

history_log = tester.run()
print(history_log)

import gc
del tester
gc.collect()
if device.startswith("cuda"):
    torch.cuda.empty_cache()
elif device.startswith("mps"):
    torch.mps.empty_cache()
```
